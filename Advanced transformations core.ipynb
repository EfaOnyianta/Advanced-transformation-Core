## Cleaning Dataframe

```python
hero_df['Hero|Publisher']. str.split('|', expand = True)
```

#### Split hero | publisher into 2 separate columns

```python
hero_df[['Hero Name', 'Publisher']] = hero_df['Hero|Publisher']. str.split('|', expand = True)
hero_df.head()
```

```python
#drop the original columan to avoid duplicated data
hero_df = hero_df.drop(columns = ['Hero|Publisher'])
hero_df.head()
```

```python
#getting rid of the brackets in the measurement column
replace = ['{','}']
for char in replace:
    hero_df['Measurements']= hero_df['Measurements'].str.replace(char,'', regex = False)
    
```

```python
hero_df.head()
#check
```

```python
height_weight = hero_df['Measurements'].apply(pd.Series)
height_weight
```

```python
#split measurements into height/ weight. indicate to split on the , into 2 columns 
hero_df[['Height(cm)','Weight(kg)']] = hero_df['Measurements'].str.split(',', expand = True)
hero_df.head()
```

```python
hero_df = hero_df.drop(columns = ['Measurements'])
hero_df.head()
#drop measurements
```

```python
#get rid of square brackets for height and weight
hero_df['Height(cm)'] = hero_df['Height(cm)'].astype(str).str.replace("[']", "", regex=True)
```

```python
hero_df['Weight(kg)'] = hero_df['Weight(kg)'].astype(str).str.replace("[']", "", regex=True)
```

```python
#get rid of any random commas in the df
hero_df.columns = hero_df.columns.str.replace(' ', '')
```

```python
hero_df['Height(cm)'].head()
```

```python
#remove extra  words from both columns
hero_df['Height(cm)'] = hero_df['Height(cm)'].str[7:]
hero_df['Height(cm)']= hero_df['Height(cm)'].str.replace('cm','', regex = False)
```

```python
hero_df['Height(cm)'] = hero_df['Height(cm)'].astype(float)
hero_df['Height(cm)']
```

```python
hero_df['Weight(kg)'] = hero_df['Weight(kg)'].str[8:]
hero_df['Weight(kg)']= hero_df['Weight(kg)'].str.replace('kg','', regex = False)
```

```python
hero_df
```

```python
hero_df['Weight(kg)'] = hero_df['Weight(kg)'].astype(float)
hero_df.info()
```
